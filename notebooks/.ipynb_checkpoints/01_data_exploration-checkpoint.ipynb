```python
# Closed-Loop Capacity Planning
# Step 1: Data Exploration

import pandas as pd
import matplotlib.pyplot as plt

print("Notebook ready")
```

```python
import os

data_path = "../data"
for root, dirs, files in os.walk(data_path):
    print(root, "->", files)
```

```python
import numpy as np
import pandas as pd

np.random.seed(42)

n_days = 365
time = pd.date_range("2023-01-01", periods=n_days, freq="D")

# base demand with weekly + seasonal pattern
demand = (
    100
    + 20 * np.sin(2 * np.pi * time.dayofyear / 365)
    + 10 * np.sin(2 * np.pi * time.dayofweek / 7)
    + np.random.normal(0, 5, n_days)
)

# provisioned capacity with lag
capacity = np.roll(demand, 3) + np.random.normal(0, 8, n_days)
capacity[:3] = demand[:3] + np.random.normal(0, 5, 3)

# SLA violation when demand > capacity
sla_violation = (demand > capacity).astype(int)

# cost proportional to overprovisioning
cost = np.maximum(capacity - demand, 0) * 0.5

df = pd.DataFrame({
    "date": time,
    "demand": demand,
    "capacity": capacity,
    "sla_violation": sla_violation,
    "cost": cost
})

df.head()
```

```python
df.to_csv("../data/raw/capacity_timeseries.csv", index=False)
print("Saved to data/raw/")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(df["date"], df["demand"], label="Demand")
plt.plot(df["date"], df["capacity"], label="Capacity")
plt.legend()
plt.title("Closed-Loop Capacity System")
plt.show()
```

```python
print("SLA violation rate:", df["sla_violation"].mean())
print("Average cost:", df["cost"].mean())
```

```python
from sklearn.linear_model import LinearRegression

df["lag1"] = df["demand"].shift(1)
df["lag2"] = df["demand"].shift(2)
df["lag3"] = df["demand"].shift(3)
df = df.dropna()

X = df[["lag1","lag2","lag3"]]
y = df["demand"]

model = LinearRegression()
model.fit(X, y)

df["forecast"] = model.predict(X)
```

```python
df[["demand", "forecast"]].head()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df["date"], df["demand"], label="Actual Demand")
plt.plot(df["date"], df["forecast"], label="Forecast")
plt.legend()
plt.title("Forecast vs Actual Demand")
plt.show()
```

```python
df["smart_capacity"] = df["forecast"] * 1.05  # 5% safety buffer

df["smart_sla"] = (df["demand"] > df["smart_capacity"]).astype(int)
df["smart_cost"] = np.maximum(df["smart_capacity"] - df["demand"], 0) * 0.5

print("OLD SLA rate:", df["sla_violation"].mean())
print("NEW SLA rate:", df["smart_sla"].mean())

print("OLD cost:", df["cost"].mean())
print("NEW cost:", df["smart_cost"].mean())
```

```python
plt.figure(figsize=(12,5))
plt.plot(df["date"], df["demand"], label="Demand")
plt.plot(df["date"], df["capacity"], label="Old Capacity")
plt.plot(df["date"], df["smart_capacity"], label="Smart Capacity")
plt.legend()
plt.title("Closed-Loop Capacity Optimization")
plt.show()
```

```python
#Phase 1 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Make a clean working copy
df2 = df.copy()

# Create lag features (based ONLY on past demand)
df2["lag1"] = df2["demand"].shift(1)
df2["lag2"] = df2["demand"].shift(2)
df2["lag3"] = df2["demand"].shift(3)

# Drop rows where lags are missing (first 3 rows)
df2 = df2.dropna().reset_index(drop=True)

df2.head()
```

```python
split_idx = int(len(df2) * 0.70)

train = df2.iloc[:split_idx].copy()
test  = df2.iloc[split_idx:].copy()

print("Train rows:", len(train), "Test rows:", len(test))
print("Train date range:", train["date"].min(), "to", train["date"].max())
print("Test date range :", test["date"].min(),  "to", test["date"].max())
```

```python
X_train = train[["lag1", "lag2", "lag3"]]
y_train = train["demand"]

X_test  = test[["lag1", "lag2", "lag3"]]
y_test  = test["demand"]

model = LinearRegression()
model.fit(X_train, y_train)

test["forecast"] = model.predict(X_test)

test[["date", "demand", "forecast"]].head()
```

```python
# OLD policy uses your existing "capacity" column
old_sla  = (test["demand"] > test["capacity"]).mean()
old_cost = np.maximum(test["capacity"] - test["demand"], 0).mean() * 0.5

# NEW policy uses forecast + buffer
buffer = 1.05
test["smart_capacity"] = test["forecast"] * buffer

new_sla  = (test["demand"] > test["smart_capacity"]).mean()
new_cost = np.maximum(test["smart_capacity"] - test["demand"], 0).mean() * 0.5

print("=== TEST-ONLY RESULTS ===")
print("OLD SLA rate:", old_sla)
print("NEW SLA rate:", new_sla)
print("OLD cost    :", old_cost)
print("NEW cost    :", new_cost)
```

```python
plt.figure(figsize=(12,5))
plt.plot(test["date"], test["demand"], label="Demand")
plt.plot(test["date"], test["capacity"], label="Old Capacity")
plt.plot(test["date"], test["smart_capacity"], label="Smart Capacity")
plt.legend()
plt.title("TEST Window: Demand vs Capacity Policies")
plt.show()
```

```python
#Phase 2
buffers = np.linspace(1.0, 1.3, 15)

results = []

for b in buffers:
    test["smart_capacity_b"] = test["forecast"] * b

    sla = (test["demand"] > test["smart_capacity_b"]).mean()
    cost = np.maximum(test["smart_capacity_b"] - test["demand"], 0).mean()

    results.append((b, sla, cost))

tradeoff = pd.DataFrame(results, columns=["buffer", "sla_rate", "cost"])
tradeoff
```

```python
plt.figure(figsize=(8,6))
plt.plot(tradeoff["cost"], tradeoff["sla_rate"], marker="o")
plt.xlabel("Average Overprovisioning Cost")
plt.ylabel("SLA Violation Rate")
plt.title("Cost vs SLA Tradeoff Curve")
plt.gca().invert_yaxis()
plt.grid()
plt.show()
```

```python
best = tradeoff.loc[tradeoff["sla_rate"].idxmin()]
best
```

```python
plt.figure(figsize=(8,6))
plt.plot(tradeoff["cost"], tradeoff["sla_rate"], marker="o")
plt.scatter(best["cost"], best["sla_rate"], s=120, color="red")
plt.xlabel("Average Overprovisioning Cost")
plt.ylabel("SLA Violation Rate")
plt.title("Pareto Frontier: Cost vs SLA")
plt.gca().invert_yaxis()
plt.grid()
plt.show()
```
